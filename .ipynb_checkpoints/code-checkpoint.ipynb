```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('weight-height.csv')
```

```python
data.head()
```

```python
y=data['Weight']
N=10000
z=y[y.idxmax()]-y[y.idxmin()]
r=y/z
r.head()
```

```python
x=data['Height']
w=(x[x.idxmax()])-(x[x.idxmin()])
q=x/w
q.head()
```

```python
m=0
c=0
X=np.array([q])
Y=np.array([r])
X.reshape(1,10000)
Y.reshape(1,10000)
```

```python
plt.plot(X,Y,'go')

plt.ylabel('Data')
plt.show()
```

```python
for i in range(10000):
    dm = (2.0/N) *  -np.sum(np.multiply(X,(Y - (m*X +c))))
   
    dc = (2.0/N) * -np.sum(Y - (m*X + c))
    m = m -0.001 * dm
    c = c - 0.001 * dc
  
```
